## KNN k-nearest neighbor 

```python
import pandas as pd
df = pd.read_csv("diabetes.csv")
df
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
y=df['Outcome']
# x=df.iloc[:,0:8]
x=df.drop('Outcome',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
nn = KNeighborsClassifier(n_neighbors=15)
model=nn.fit(x_train,y_train)
pred = model.predict(x_test)
pred
```

```python
                                            #OR
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=15)
model.fit(x_train,y_train)
pred = model.predict(x_test)
pred
```

```python
diff= pd.DataFrame({"Actual":y_test,"Predicted":pred})
diff
```

```python
diff.to_excel("diabetes_data.xlsx")
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,pred)
cm
```

```python
TN=cm[0][0]
FP=cm[0][1]
FN=cm[1][0]
TP=cm[1][1]
print(TN,FP,FN,TP)
```

# Accuracy Score

```python
print("Accuracy_Score: ",(TP+TN)/(TN+TP+FN+FP))
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy Score: ",accuracy_score(y_test,pred))
```

## ERROR RATE

```python
print("error rate: ",(FP+FN)/(TN+TP+FN+FP))
```

```python
print("error rate: ",1-accuracy_score(y_test,pred))
```

## Sensitivity & Specificity

```python
print("Sensitivity: ",TP/(TP+FN)) # Positive Accuracy
```

```python
from sklearn.metrics import recall_score
print("Sensitivity: ",recall_score(y_test,pred))
```

```python
print("Specificity: ",TN/(TN+FP)) #Negative Accuracy 
```

```python
print("Specificity: ",recall_score(y_test,pred,pos_label=0))
```

## PRECISION Score

```python
print("precision score: ",TP/(TP+FP))
```

```python
from sklearn.metrics import precision_score
print("precision_score: ",precision_score(y_test,pred))
```

```python
from sklearn.metrics import accuracy_score
accuracy=[]
for i in range(1,31):
    nn = KNeighborsClassifier(n_neighbors=i)
    model=nn.fit(x_train,y_train)
    pred = model.predict(x_test)
    accuracy.append(accuracy_score(y_test,pred))
# accuracy
print(max(accuracy))   
    
    
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,2))
plt.plot(range(1,31),accuracy,marker="D")
plt.grid(True)
plt.show()
```

```python
from sklearn import metrics
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm , display_labels=['No diabetes','diabetes'])
cm_display.plot()
plt.show()
```

## Random Forest

```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df.isna().sum()
```

```python
import numpy as np
df['Sex']=np.where(df['Sex']=='male',1,0)
```

```python
df
```

```python
df.columns
```

```python
x=df[['Pclass','Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
y=df['Survived']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(criterion='entropy',n_estimators=100,random_state=1)
model = rf.fit(x_train,y_train)
pred=model.predict(x_test)
pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,pred)
cm
```

```python
print("Accuracy_score",accuracy_score(y_test,pred))
```

```python
from sklearn.metrics import recall_score
print("Sensitivity: ",recall_score(y_test,pred))
```

```python
print("Specificity: ",recall_score(y_test,pred,pos_label=0))
```

```python
from sklearn.metrics import precision_score
print("precision_score: ",precision_score(y_test,pred))
```

```python
sample = x_test.iloc[0:1]
print(sample)
prediction=model.predict(sample)
print(f"Predicted survival: {'survived' if prediction[0]==1 else 'Not survived'}")
```

```python
sample_dict=sample.iloc[0].to_dict()
print(f"Sample Passenger: {sample_dict}")
print(f"Predicted survival: {'survived' if prediction[0]==1 else 'Not survived'}")
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
choose_tree=model.estimators_[0]
plt.figure(figsize=(15,10))
plot_tree(choose_tree,
         feature_names=['Pclass','Sex', 'Age', 'SibSp', 'Parch', 'Fare'],
         class_names=['Not survived','Survived'],
         filled=True,rounded=True)
plt.show()
```
