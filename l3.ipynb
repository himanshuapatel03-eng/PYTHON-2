```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
soup = BeautifulSoup(open('Popular_Quotes.html','r',encoding='utf-8'))

quote = soup.find_all('div',class_='quoteText')
quotes = []
author = soup.find_all('span',class_='authorOrTitle')
authors = []
like = soup.find_all('div',class_='right')
likes = []

for (i,j,k) in zip(quote,author,like):
    quotes.append(i.text.split('―')[0].strip()[1:-1])
    authors.append(j.text.strip().split(',')[0])
    likes.append(k.text.strip()[0:-6])

df = pd.DataFrame({'Quotes':quotes,'Authors':authors,'Likes':likes})
df
```

```python
def get_coordinates():
    # api_key = input('Enter API Key: ')
    api_key = API_Key
    city = input('Enter City Name: ')
    request_url = f'http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}'
    response = requests.get(request_url)
    coordinates = response.json()
    # print(json.dumps(coordinates,indent=5))
    lattitude = coordinates[0]['lat']
    longitude = coordinates[0]['lon']
    return lattitude,longitude,city

lat,lon,city = get_coordinates()

print(city)
print('--------')
print('Lattitude: ',lat)
print('Logitude: ',lon)
```

```python
def current_weather_data():
    # api_key = input('Enter API Key: ')
    api_key = API_Key
    request_url = f'https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}'
    response = requests.get(request_url)
    current_weather = response.json()
    print(json.dumps(current_weather,indent=5))
    print('----------')
    print('Weather:',current_weather['weather'])
    print('----------')
    print('Description:',current_weather['weather'][0]['description'])
    print('----------')
    print('Temperature:',current_weather['main']['temp'])
    print('----------')
    print('Pressure:',current_weather['main']['pressure'])
    print('----------')
    print('Humidity:',current_weather['main']['humidity'])
    print('----------')
    print('Wind Speed:',current_weather['wind']['speed'])
    print('----------')
    print('Visibilty:',current_weather['visibility'])

current_weather_data()
```
