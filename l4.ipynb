```python
API_Key = '8a1c4f528052ff9f2e1b1afe26e4cb8b'
import requests,json
import pandas as pd
```

```python
def get_coordinates():
    # api_key = input('Enter API Key: ')
    api_key = API_Key
    city = input('Enter City Name: ')
    request_url = f'http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}'
    response = requests.get(request_url)
    coordinates = response.json()
    # print(json.dumps(coordinates,indent=5))
    lattitude = coordinates[0]['lat']
    longitude = coordinates[0]['lon']
    return lattitude,longitude,city

lat,lon,city = get_coordinates()

print(city)
print('--------')
print('Lattitude: ',lat)
print('Logitude: ',lon)
```

```python
def get_air_pollution():
    api_key = API_Key
    url = f'http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={API_Key}'
    response = requests.get(url)

    pollution = response.json()

    a = pollution['list'][0]['main']['aqi']

    if(a==1):
        print('GOOD')
    if(a==2):
        print('FAIR')
    if(a==3):
        print('MODERATE')
    if(a==4):
        print('POOR')
    if(a==5):
        print('VERY POOR')

get_air_pollution()
```

```python
dt = []
des = []
temp = []
pressure = []
hum = []
visibility = []
ws = []

def forecast5():
    url = f'https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={API_Key}'
    response = requests.get(url)
    
    f = response.json()
    # print(json.dumps(f,indent=5))

    for i in range(0,40):
        dt.append(f['list'][i]['dt_txt'])
        des.append(f['list'][i]['weather'][0]['description'])
        temp.append(f['list'][i]['main']['temp'])
        pressure.append(f['list'][i]['main']['pressure'])
        hum.append(f['list'][i]['main']['humidity'])
        visibility.append(f['list'][i]['visibility'])
        ws.append(f['list'][i]['wind']['speed'])
forecast5()

df = pd.DataFrame({'Date':dt,'Description':des,'Temperature':temp,'Pressure':pressure,'Humidity':hum,'Visibility':visibility,'Wind Speed':ws})
df
```
